# Table with Matplotlib

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Setting seaborn as default style even if use only matplotlib
sns.set()
```

```python
fig, axes = plt.subplots(1, 3)
fig.suptitle('1 row x 2 columns axes with no data')
```

* Now __axes__ is an array of __AxesSubplot__, so we can access each __ax__ separetely and set a different title, for instance.
* More arguments:
  * _figsize_ set the total dimension of our figure
  * _sharex_ and sharey are used to share one or both axes between the charts (needed data to work)

```python
fig, axes = plt.subplots(3, 4, sharex=True, figsize=(16,8))
fig.suptitle('3 rows x 4 columns axes with no data')
```

## Example 1: Simple Table

```python
data = [[1, 2, 3, 4, 5],
        [10,20,30,40,50],
        [11,21,31,41,51]]

plt.table(data, loc='center', colLabels=['A','B','C','D','E'])

plt.show()
```

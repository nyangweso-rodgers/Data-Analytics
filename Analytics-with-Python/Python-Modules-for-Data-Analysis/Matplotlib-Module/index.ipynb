```python
# import the packages
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Plot Single Graph
* When you want to draw a single graph, you can use __plt.plot()__ directly, but using __plt.subplots()__ is also OK.

### Using plt.plot()

```python
plt.plot(np.random.rand(10))  # 10 random number between [0, 1)
plt.title("My Title")
plt.show()
```

### Using plt.subplots()

```python
fig, ax = plt.subplots()  # Create a figure and a set of subplots.

ax.plot(np.random.rand(10))
ax.set_title("My Title")
plt.show()
```

## Plot Multiple Graphs
* When you want to draw multiple graphs, you should use __plt.subplots()__.

```python
# 2x2, a total of 4 graphs
n_rows = 2
n_cols = 2

# Create a figure and subplots of size 2x2
fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols)  

for row_num in range(n_rows):
    for col_num in range(n_cols):
        ax = axes[row_num][col_num]
        ax.plot(np.random.rand(10))
        
fig.suptitle("main title")
fig.tight_layout()  # Smaller padding
plt.show()
```
